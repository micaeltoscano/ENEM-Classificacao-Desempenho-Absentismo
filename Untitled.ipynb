```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_parquet("enem_parquet/2019")
```

```python
colunas_para_excluir = ['NU_INSCRICAO',  'TP_SEXO',  'TP_COR_RACA', 'TP_NACIONALIDADE', 'TP_ANO_CONCLUIU', 
                        'CO_MUNICIPIO_ESC', 'NO_MUNICIPIO_ESC', 'CO_UF_ESC', 'SG_UF_ESC',
                        'TP_SIT_FUNC_ESC',
                        'CO_MUNICIPIO_PROVA', 'NO_MUNICIPIO_PROVA', 'CO_UF_PROVA',
                        'SG_UF_PROVA', 'TP_STATUS_REDACAO',
                        'NU_NOTA_COMP1', 'NU_NOTA_COMP2', 'NU_NOTA_COMP3', 'NU_NOTA_COMP4',
                        'NU_NOTA_COMP5', 'NU_NOTA_REDACAO']

df = df.drop(columns=colunas_para_excluir)
```

```python
FALTOU = (
    (df['TP_PRESENCA_CH'] == 0) & 
    (df['TP_PRESENCA_LC'] == 0) & 
    (df['TP_PRESENCA_CN'] == 0) & 
    (df['TP_PRESENCA_MT'] == 0)
)

df['FALTOU'] = FALTOU.astype(int)
```

```python
df = df[df['FALTOU'] == 0]
```

```python
prova_map = {
    'CN': {503:'Azul', 504:'Amarela', 505:'Cinza', 506:'Rosa', 519:'Azul', 523:'Amarela PPL'},
    'CH': {507:'Azul', 508:'Amarela', 509:'Cinza', 510:'Rosa', 520:'Azul', 524:'Amarela PPL'},
    'LC': {511:'Azul', 512:'Amarela', 513:'Cinza', 514:'Rosa', 521:'Azul', 525:'Amarela PPL'},
    'MT': {515:'Azul', 516:'Amarela', 517:'Cinza', 518:'Rosa', 522:'Azul', 526:'Amarela PPL'}
}
df['COR_CN'] = df['CO_PROVA_CN'].map(prova_map['CN'])
df['COR_CH'] = df['CO_PROVA_CH'].map(prova_map['CH'])
df['COR_LC'] = df['CO_PROVA_LC'].map(prova_map['LC'])
df['COR_MT'] = df['CO_PROVA_MT'].map(prova_map['MT'])
```

```python
gabaritos = {}

for area in ['CN', 'CH', 'LC', 'MT']:
    gabaritos[area] = (
        df[[f'COR_{area}', f'TX_GABARITO_{area}']]
        .dropna()
        .drop_duplicates()
        .set_index(f'COR_{area}')[f'TX_GABARITO_{area}']
        .to_dict()
    )

print(gabaritos)
```

```python
INVALIDOS = {'*', '.', '9'}

def gabarito_por_questao(respostas, gabarito):
    if pd.isna(respostas) or pd.isna(gabarito):
        return [None] * len(gabarito)
    
    return [
        None if r in INVALIDOS or g in INVALIDOS else int(r == g)
        for r, g in zip(respostas, gabarito)
    ]
```

```python
INVALIDOS = {'9','*','.'}

n = len(df)
matriz = np.full((n,180), np.nan, dtype=np.float32)

questao_idx = 0

for area in ['CN','CH','LC','MT']:

    gabs = df[f'COR_{area}'].map(gabaritos[area]).values
    resps = df[f'TX_RESPOSTAS_{area}'].values

    for q in range(45):

        r = np.array([s[q] if isinstance(s,str) else None for s in resps])
        g = np.array([s[q] if isinstance(s,str) else None for s in gabs])

        valido = (
            (r != None) &
            (g != None) &
            (~np.isin(r,list(INVALIDOS))) &
            (~np.isin(g,list(INVALIDOS)))
        )

        matriz[valido,questao_idx] = (r[valido] == g[valido])

        questao_idx += 1
```

```python
colunas = [f'Q{i:3d}' for i in range(1, 181)]
df_acertos = pd.DataFrame(matriz, columns = colunas)
df_acertos = df_acertos.fillna(-1)
```

```python
df_final = pd.concat(
    [df_acertos.reset_index(drop=True),
     df.reset_index(drop=True)],
    axis=1
)
```
