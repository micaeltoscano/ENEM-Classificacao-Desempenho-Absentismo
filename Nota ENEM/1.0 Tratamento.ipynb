# Random Forest na Classificação do Desempenho dos Alunos

# Seção 1: Importação do Dataset e Dataset Usado

```python
import pandas as pd
import string
```

```python
df = pd.read_parquet("Dados Limpos")
```

# Seção 2: Tratamento de Dados e Criação de Features

```python
def transformar_colunas(df):

    colunas = [
    'Q001','Q002','Q003','Q004','Q006','Q007','Q008','Q009','Q010',
    'Q011','Q012','Q013','Q014','Q015','Q016','Q017','Q018',
    'Q019','Q020','Q021', 'Q022','Q023','Q024','Q025'
    ]

    letras = {l:i for i,l in enumerate(string.ascii_uppercase)}

    for col in colunas:

        df[col] = df[col].map(letras)

        df[col] = df[col].fillna(-1)

    return df
```

```python
df = transformar_colunas(df)
```

```python
df['MEDIA'] = (df['NU_NOTA_CN'] + df['NU_NOTA_CH'] + df['NU_NOTA_MT']+  df['NU_NOTA_LC'] + df['NU_NOTA_REDACAO']) / 5

mediana = df['MEDIA'].median()
df['CLASSE_NOTA'] = (df['MEDIA'] < mediana).astype(int)
```

```python
df['CLASSE_NOTA'].value_counts()
```

```python
df['ESCOLARIDADE_PAIS'] = df['Q001'] + df['Q002']

df['INDICE_SOCIO'] = (
    df['Q006']*2 +
    df['Q024'] +
    df['Q010'] +
    df['Q008']
)

df['CAPITAL_EDUCACIONAL'] = (
    df['Q001'] +
    df['Q002'] +
    df['Q006']
)


df['RECURSOS_ESTUDO'] = (
    df['Q024'] +
    df['Q025'] +
    df['Q022']
)

df['TIPO_ESCOLA_COMP'] = (
    df['TP_ESCOLA']*10 +
    df['TP_DEPENDENCIA_ADM_ESC']
)
```

```python
df.to_parquet("dados_nota_enem.parquet")
```
